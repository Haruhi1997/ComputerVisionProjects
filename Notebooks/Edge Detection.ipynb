```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2


img = cv2.imread('imageFileName.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.imshow(laplacian,cmap='gray')
plt.show()
```

```python
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('doggie.jpg',0)
edges = cv2.Canny(img,0,200)
print(type(edges),edges.shape)
inverted = (255-edges)
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.imshow(inverted,cmap = 'gray')
plt.show()
```

```python
inverted_grayscale = 255-img
#Applying Gaussian blur
img_blur = cv2.GaussianBlur(inverted_grayscale,ksize=(21,21),sigmaX=0,sigmaY=0)
plt.imshow(img_blur,cmap='gray')
plt.show()
```

```python
def dodgeV2(image, mask):
  return cv2.divide(image, 255-mask, scale=256)
```

```python
img_blend = dodgeV2(img,img_blur)
plt.imshow(img_blend,cmap='gray')
plt.show()
cv2.imwrite('doggiesketch.jpg',img_blend)
```

# Real time image capture and conversion to sketch

```python

import cv2
camera = cv2.VideoCapture(0)
cv2.namedWindow("testcapture")
img_counter=0
while True:
    ret,frame= camera.read()
    if not ret:
        print('Failed to grab frame')
        break
    cv2.imshow('test',frame)
    #input('Press 1 to quit or space to capture image')
    k=cv2.waitKey(1)
    if k%256 == 27:
        print("Closing..")
        break
    elif k%256==32:
        #apturing img
        img_name = "opencv"+str(img_counter)+".png"
        cv2.imwrite(img_name,frame)
        img_counter+=1
        img = cv2.imread(img_name)
        edges = cv2.Canny(img,0,200)
        inverted = (255-edges)
        inverted_grayscale = 255-img
        img_blur = cv2.GaussianBlur(inverted_grayscale,ksize=(21,21),sigmaX=0,sigmaY=0)
        img_blend = dodgeV2(img,img_blur)
        cv2.imwrite("sketch"+str(img_counter)+".jpg",img_blend)
camera.release()
cv2.destroyAllWindows()
    
```

# Applying Sobel filters on grid

```python
sobel_horizontal = np.array([1,2,1]),np.array([0,0,0]),np.array([-1,-2,-1])
sobel_vertical = np.array([-1,0,1]),np.array([-2,0,2]),np.array([-1,0,1])
grid = cv2.imread('grid.png')
from scipy import ndimage
from skimage.color import rgb2gray
```

```python

#Convert to grayscale for easy understanding
grid_gs= rgb2gray(grid)

sobel_h = ndimage.convolve(grid_gs,sobel_horizontal,mode='reflect')
sobel_v = ndimage.convolve(grid_gs,sobel_vertical,mode='reflect')
plt.imshow(sobel_h,cmap='gray')
```

```python
plt.imshow(sobel_v,cmap='gray')
```

```python
#Kernel laplace can detect both horizontal and vertical edges at the same time

laplace_kernel = np.array([1,1,1]),np.array([1,-8,1]),np.array([1,1,1])
laplace_img = ndimage.convolve(grid_gs,laplace_kernel,mode='reflect')
plt.imshow(laplace_img,cmap='gray')
```

## K-NN for semantic segmentation

```python
#Dividing by 255 to bring all pixel values between 0-1
img = cv2.imread('doggie.jpg')/255
print("Original shape: ",img.shape)
#Reshape width and height to one array
img_reshape = img.reshape(img.shape[0]*img.shape[1],img.shape[2])
print("Reshaped image: ",img_reshape.shape)
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5,random_state=0).fit(img_reshape)
result = km.cluster_centers_[km.labels_]
```

```python
back_to_orig_shape= result.reshape(img.shape[0],img.shape[1],img.shape[2])
plt.imshow(back_to_orig_shape)
```
